## Check Annotation Progress

```python
csv_path = 'data/training/annotations2.csv'

if not os.path.exists(csv_path):
    print(f"File not found: {csv_path}")
else:
    df = pd.read_csv(csv_path)
    total = len(df)
    
    # Count non-empty captions
    df['caption'] = df['caption'].astype(str)
    annotated = df[(df['caption'] != 'nan') & (df['caption'].str.strip() != '')].shape[0]
    remaining = total - annotated
    
    # Progress bar
    progress = int((annotated / total) * 50) if total > 0 else 0
    bar = '█' * progress + '░' * (50 - progress)
    
    print(f"Total images: {total}")
    print(f"Annotated: {annotated} ({annotated/total*100:.1f}%)")
    print(f"Remaining: {remaining}")
    print(f"\n[{bar}] {annotated/total*100:.1f}%")
    
    if annotated > 0:
        print("\nSample annotations:")
        completed = df[(df['caption'] != 'nan') & (df['caption'].str.strip() != '')]
        for idx, row in completed.head(3).iterrows():
            print(f"  {row['filename']}: {row['caption'][:60]}...")
    
    if remaining == 0:
        print("\nStatus: Complete - ready for fine-tuning")
```
